```python
# Load the needed packages
from glob import glob
import matplotlib.pyplot as plt

from awot.AirborneData import AirborneData
from awot.graph.radar_vertical import RadarSweepPlot

%matplotlib inline
```

```python
# We'll first show an example of plotting a cfradial formatted file from NOAA P-3 (2012)
fnc = '/Users/nickguy/data/p3radar/HRD_test/20120828/cfradial/cfrad.20120828_120541.113_to_20120828_120547.113_N42R_v1_s00_az-19.48_AIR.nc'

RadName = 'NOAA P-3 Tail Radar'

Tx = 'NOAA P-3 Reflectivity'

Field = 'DBZ'
FLim = (0.,60.)
#Field = 'VEL'
#FLim = (-40.,40.)
RngRings = [15.,30.,46.,60.]
RngLim = (-70.,70.)
HtLim = (-10.,20.)
dMax = 45.

# Creating axes outside seems to screw up basemap
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

# Initiate an Airborne class instance
fl = AirborneData()

# Get the tail radar data
fl.get_radar_data(fname=fnc, platform='p-3', file_format='netcdf', 
                  instrument='tdr_sweep')
# Create a RadarSweep
rgp = RadarSweepPlot(fl, instrument='tdr_sweep')

# The most common way to view the data is to use the track relative plot which projects the 
# data onto a coordinate system about the aircraft track (removes tilt angles)
rgp.plot_track_relative(Field, mask_procedure=None, mask_tuple=None,
      cb_pad=.02, cb_orient='vertical', ylims=(-10.,15.),
      ax=ax, fig=fig)

# Notice that a warning is issued about not having a basemap set.  This does not affect the plot.
```
