```python
# This script produces a 4-panel plot from a gridded data file produced
# by the windsyn analysis program
```

```python
# Load the needed packages
from glob import glob
import matplotlib.pyplot as plt

from awot.AirborneData import AirborneData
from awot.graph.radar_horizontal import RadarHorizontalPlot
from awot.graph.flight_level import FlightLevel

%matplotlib inline
```

```python
# Set the project name
Project="DYNAMO"

# Choose what file to process
yymmdd, modn = '111124', '0705'

# Set the data directory
fDir = "/Users/nickguy/data/dynamo/"+yymmdd+"I/"

# Construct the full path name for windsyn NetCDF file
P3Radf = str(glob(fDir+modn+"/*"+modn+"*windsyn*.nc")).strip('[]')
# Construct the full path name for Flight level NetCDF file
FltLevf = str(glob(fDir+"20*"+yymmdd+"*_DJ*.nc")).strip('[]')

# Optional settings
start_time = "2011-11-24 07:05:00"
end_time = "2011-11-24 07:45:00"
corners = [78.5, -0.8, 80.1, 0.8]

dLon = 0.5
dLat = 0.5

figtitle = '24 Nov RCE'
```

```python
# Set up some characteristics for plotting

# Set map projection to use
proj = 'cea'

Wbarb_Spacing = 300 # Spacing of wind barbs along flight path (sec)

# Choose the X-axis time step (in seconds) where major labels will be
XlabStride = 60

# Should landmarks be plotted? [If yes, then modify the section below
Lmarks=False
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(7, 7))#, sharex='col', sharey='row')

# Initiate an Airborne class instance
fl1 = AirborneData()
fl2 = AirborneData()
fl3 = AirborneData()
fl4 = AirborneData()

# Get the airborne data
fl1.get_flight_data(fname=FltLevf[1:-1], platform='p-3', file_format='netcdf')
fl2.get_flight_data(fname=FltLevf[1:-1], platform='p-3', file_format='netcdf')
fl3.get_flight_data(fname=FltLevf[1:-1], platform='p-3', file_format='netcdf')
fl4.get_flight_data(fname=FltLevf[1:-1], platform='p-3', file_format='netcdf')
# Get the tail radar data
fl1.get_radar_data(fname=fDir+modn+"/"+modn, platform='p-3', file_format='binary', 
                  instrument='tdr_grid')
fl2.get_radar_data(fname=fDir+modn+"/"+modn, platform='p-3', file_format='binary', 
                  instrument='tdr_grid')
fl3.get_radar_data(fname=fDir+modn+"/"+modn, platform='p-3', file_format='binary', 
                  instrument='tdr_grid')
fl4.get_radar_data(fname=fDir+modn+"/"+modn, platform='p-3', file_format='binary', 
                  instrument='tdr_grid')

# Set the map for plotting
fl1.create_basemap(corners=corners, proj=proj, resolution='l', area_thresh=1., 
                  dLat=dLat, dLon=dLon, ax=ax1)
fl2.create_basemap(corners=corners, proj=proj, resolution='l', area_thresh=1., 
                  dLat=dLat, dLon=dLon, ax=ax2)
fl3.create_basemap(corners=corners, proj=proj, resolution='l', area_thresh=1., 
                  dLat=dLat, dLon=dLon, ax=ax3)
fl4.create_basemap(corners=corners, proj=proj, resolution='l', area_thresh=1., 
                  dLat=dLat, dLon=dLon, ax=ax4)

# Create a RadarGrid
rgp1 = RadarHorizontalPlot(fl1, instrument='tdr_grid')
rgp1.plot_ppi('dBZ', ppi_height=2., vmin=15., vmax=60., title=' ',
              color_bar=True, cb_pad="10%", cb_loc='bottom', cb_tick_int=4,
              ax=ax1)
rgp2 = RadarHorizontalPlot(fl2, instrument='tdr_grid')
rgp2.plot_ppi('Wwind', ppi_height=2., vmin=-5., vmax=5., title=' ',
              mask_procedure='inside', mask_tuple=('Wwind', -1., 1.),
              cmap='RdBu_r',
              color_bar=True, cb_pad="10%", cb_loc='bottom', 
              ax=ax2)
rgp3 = RadarHorizontalPlot(fl3, instrument='tdr_grid')
rgp3.plot_ppi('dBZ', ppi_height=5., vmin=15., vmax=60., title=' ',
              color_bar=True, cb_pad="10%", cb_loc='bottom', cb_tick_int=4,
              ax=ax3)
rgp4 = RadarHorizontalPlot(fl4, instrument='tdr_grid')
rgp4.plot_ppi('Wwind', ppi_height=5., vmin=-5., vmax=5., title=' ',
              mask_procedure='inside', mask_tuple=('Wwind', -1., 1.),
              cmap='RdBu_r',
              color_bar=True, cb_pad="10%", cb_loc='bottom', 
              ax=ax4)

# Create a Flightlevel instance for the track
flp1 = FlightLevel(fl1)
flp2 = FlightLevel(fl2)
flp3 = FlightLevel(fl3)
flp4 = FlightLevel(fl4)

flp1.plot_trackmap(
              start_time=start_time, end_time=end_time,
#             color_by_altitude=True, track_cmap='spectral',
#              min_altitude=50., max_altitude= 8000.,
              addlegend=False, addtitle=False, ax=ax1)
flp2.plot_trackmap(
              start_time=start_time, end_time=end_time,
#              min_altitude=50., max_altitude= 8000.,
              addlegend=False, addtitle=False, ax=ax2)
flp3.plot_trackmap(
              start_time=start_time, end_time=end_time,
#              min_altitude=50., max_altitude= 8000.,
              addlegend=False, addtitle=False, ax=ax3)
flp4.plot_trackmap(
              start_time=start_time, end_time=end_time,
#              min_altitude=50., max_altitude= 8000.,
              addlegend=False, addtitle=False, ax=ax4)

#flp.draw_scale(location='lower_middle')
#flp.draw_barbs(barbspacing=Wbarb_Spacing)
        
flp1.time_stamps(labelspacing=600,
              start_time=start_time, end_time=end_time, ax=ax1)
flp1.time_stamps(labelspacing=600,
              start_time=start_time, end_time=end_time, ax=ax2)
flp1.time_stamps(labelspacing=600,
              start_time=start_time, end_time=end_time, ax=ax3)
flp1.time_stamps(labelspacing=600,
              start_time=start_time, end_time=end_time, ax=ax4)
```
