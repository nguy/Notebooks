```python
# This script inputs a gridded NetCDF windsyn file and produces a 2-panel plot
# of horizontal CAPPI of reflectivity at 2 km and overlays the flight track and
# wind vectors derived from the radar. 

# A cross-section along the flight track is made of the vertical wind velocity in the 1st plot.
# A cross-section along lon/lat pairs of reflectivity is shown in the 2nd plot
```

```python
# Load the needed packages
from glob import glob
import matplotlib.pyplot as plt
import numpy as np

from awot.AirborneData import AirborneData
from awot.graph.radar_grid import RadarGridPlot
from awot.graph.flight_level import FlightLevel

%matplotlib inline
```

```python
# Set the project name
Project="DYNAMO"

# Choose what file to process
yymmdd, modn = '111124', '0351'

# Set the data directory
fDir = "/Users/nickguy/data/dynamo/"+yymmdd+"I/"

# Construct the full path name for windsyn NetCDF file
P3Radf = str(glob(fDir+modn+"/*"+modn+"*windsyn*.nc")).strip('[]')
# Construct the full path name for Flight level NetCDF file
FltLevf = str(glob(fDir+"20*"+yymmdd+"*_DJ*.nc")).strip('[]')

# Optional settings
start_time = "2011-11-24 03:51:00"
end_time = "2011-11-24 04:57:00"
corners = [77.8, -2.0, 79.6, -0.2]

dLon = 0.5
dLat = 0.5

figtitle = '24 Nov RCE'
```

```python
# Set up some characteristics for plotting

# Set map projection to use
proj = 'cea'

Wbarb_Spacing = 300 # Spacing of wind barbs along flight path (sec)

# Choose the X-axis time step (in seconds) where major labels will be
XlabStride = 60

# Should landmarks be plotted? [If yes, then modify the section below
Lmarks=True

if Lmarks:
    # Create a list of Landmark data
    LocMark = []
    
    # Add locations as [ StringName, Longitude, Latitude ,XlabelOffset, YlabelOffset]
    LocMark.append(['Diego Garcia', 72.4160, -7.3117, 0.1, -0.6])
    LocMark.append(['R/V Revelle', 80.5010, 0.12167, -0.4, -0.6])
    LocMark.append(['Gan', 73.1017, -0.6308, -0.9, 0.0])
    LocMark.append(['R/V Marai', 80.50, -7.98, -0.1, -0.6])
```

```python
# Build a few variables for plotting the labels
if Lmarks:
    # Build arrays for plotting
    Labels = []
    LabLons = []
    LabLats = []
    XOffset = []
    YOffset = []
    for L1, L2, L3, L4, L5 in LocMark:
        Labels.append(L1)
        LabLons.append(L2)
        LabLats.append(L3)
        XOffset.append(L4)
        YOffset.append(L5)
```

```python
fig, (axPPI, axXS) = plt.subplots(2, 1, figsize=(7, 7))

# Initiate an Airborne class instance
fl1 = AirborneData()

# Get the airborne data
fl1.get_flight_data(fname=FltLevf[1:-1], platform='p-3', file_format='netcdf')
# Get the tail radar data
fl1.get_radar_data(fname=fDir+modn+"/"+modn, platform='p-3', file_format='binary', 
                  instrument='tdr_grid')

# Set the map for plotting
fl1.create_basemap(corners=corners, proj=proj, resolution='l', area_thresh=1.,
                   dLat=dLat, dLon=dLon, ax=axPPI)

# Create a Flightlevel instance for the track
flp1 = FlightLevel(fl1)

flp1.plot_trackmap(
              start_time=start_time, end_time=end_time,
#             color_by_altitude=True, track_cmap='spectral',
              min_altitude=50., max_altitude= 8000.,
              addlegend=False, addtitle=False, ax=axPPI)


# Create a RadarGrid
rgp1 = RadarGridPlot(fl1, instrument='tdr_grid')
# Add PPI plot
rgp1.plot_ppi('dBZ', ppi_height=2., vmin=15., vmax=60., title=' ',
#rgp1.plot_ppi('Uwind', ppi_height=2., vmin=-20., vmax=20., title=' ',
#              cmap='RdBu_r',
              color_bar=True, cb_pad="10%", cb_loc='right', cb_tick_int=4,
              ax=axPPI)
rgp1.overlay_wind_vector(height_level=2., vscale=200, vtrim=6, qcolor='0.50')

flp1.plot_radar_cross_section(rgp1, 'Wwind',
              start_time=start_time, end_time=end_time,
              vmin=-3., vmax=3., title=' ',
              cmap='RdBu_r',
              color_bar=True, cb_orient='vertical', cb_tick_int=4,
              x_axis_array='time',
              ax=axXS)
```

```python
# Now let's make a vertical cross section from a slice between two lon/lat pairs
del fig, axPPI, axXS

fig, (axPPI, axXS) = plt.subplots(2, 1, figsize=(7, 7))

# Initiate an Airborne class instance
fl1 = AirborneData()

# Get the airborne data
fl1.get_flight_data(fname=FltLevf[1:-1], platform='p-3', file_format='netcdf')
# Get the tail radar data
fl1.get_radar_data(fname=fDir+modn+"/"+modn, platform='p-3', file_format='binary', 
                  instrument='tdr_grid')

# Set the map for plotting
fl1.create_basemap(corners=corners, proj=proj, resolution='l', area_thresh=1.,
                   dLat=dLat, dLon=dLon, ax=axPPI)

# Create a Flightlevel instance for the track
flp1 = FlightLevel(fl1)

flp1.plot_trackmap(
              start_time=start_time, end_time=end_time,
#             color_by_altitude=True, track_cmap='spectral',
              min_altitude=50., max_altitude= 8000.,
              addlegend=False, addtitle=False, ax=axPPI)


# Create a RadarGrid
rgp1 = RadarGridPlot(fl1, instrument='tdr_grid')
# Add PPI plot
rgp1.plot_ppi('dBZ', ppi_height=2., vmin=15., vmax=60., title=' ',
#rgp1.plot_ppi('Uwind', ppi_height=2., vmin=-20., vmax=20., title=' ',
#              cmap='RdBu_r',
              color_bar=True, cb_pad="10%", cb_loc='right', cb_tick_int=4,
              ax=axPPI)
rgp1.overlay_wind_vector(height_level=2., vscale=200, vtrim=6, qcolor='0.50')

# Add Cross-sectional line to horizontal plot
rgp1.plot_line_geo([78.3, 79.0], [-1.1, -1.5], lw=4, alpha=.8, line_style='w-',
                   label0=True, label_offset=(0.05,-0.05))

# Add the cross-section along those coordinates
rgp1.plot_cross_section('dBZ', (78.3, -1.1), (79.0, -1.5),
              vmin=15., vmax=60., title=' ',
              color_bar=False, cb_orient='vertical', cb_tick_int=4,
              ax=axXS)
```
