```python
import numpy as np
import os
from matplotlib import pyplot as plt
import pyart

%matplotlib inline
```

```python
fdir = '/Users/guy/data/nexrad/KILN/nex2/20140429'
fname = 'KILN20140429_231254_V06'
r=pyart.io.read(os.sep.join([fdir, fname]))
d=pyart.graph.RadarDisplay(r)
```

```python
print r.fields.keys()

print np.min(r.fields['velocity']['data']), np.max(r.fields['velocity']['data'])

# Create three different maskes to apply
mask_dict1 = {'mask_procedure':'outside','mask_val_1':10., 'mask_val_2':25.}
mask_dict2 = {'mask_procedure':'inside','mask_val_1':10., 'mask_val_2':25.}
mask_dict3 = {'mask_procedure':'greater_equal', 'mask_val_1':35.}
mask_dict4 = {'mask_procedure':'less', 'mask_val_1':35.}
mask_dict5 = {'mask_procedure':'equal', 'mask_val_1':25.}
mask_dict6 = {'mask_field': 'velocity', 'mask_procedure':'greater_equal',
              'mask_val_1':2.}
```

```python
# Set some common limits for direct comparisons
VMIN = 0. # Min DBZ in this case
VMAX = 55. # Max DBZ in this case
rrings = [50, 100]
pxlims = (-150.,150.)
pylims = (-150.,150.)
swp = 1
cmdbz = 'pyart_NWSRef'
cmvel = 'pyart_NWSVel'
```

## Plot with no masking

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_outside=False)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot with mask outside turned on

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_outside=True)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot with reflectivity < 10 and reflectivity > 25 masked

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_dictionary=mask_dict1)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot with reflectivity > 10 and reflectivity < 25 masked

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_dictionary=mask_dict2)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot with reflectivity >= 35 masked

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_dictionary=mask_dict3)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot with reflectivity < 35 masked

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_dictionary=mask_dict4)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot with reflectivity = 25 masked

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_dictionary=mask_dict5)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot of Doppler velocity with no masking

```python
d.plot('velocity', swp, cmap=cmvel, vmin=-30, vmax=30., mask_outside=False)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```

## Plot of reflectivity masked where velocity >= 2 

```python
d.plot('reflectivity', swp, cmap=cmdbz, vmin=VMIN, vmax=VMAX, mask_dictionary=mask_dict6)
d.plot_range_rings(rrings)
d.set_limits(pxlims, pylims)
```
