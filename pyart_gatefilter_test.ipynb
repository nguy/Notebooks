```python
import pyart
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Set file and read into pyart
#tfile = '/Users/guy/data/nexrad/KMPX/nex2/20130513/KMPX20130513_184838_V06'
tfile = '/Users/guy/data/nexrad/KILN/nex2/20140429/KILN20140430_012327_V06'

r = pyart.io.read(tfile)
r2 = pyart.io.read(tfile)
r.fields['reflectivity']['data']
```

```python
d = pyart.graph.RadarDisplay(r)
d.plot('reflectivity', 1, cmap='pyart_LangRainbow12', vmin=-10., vmax=65.)
d.set_limits((-250., 250), (-250,250))
```

```python
# Get the original mask
origmask = r.fields['reflectivity']['data'].mask
origmask
np.sum(origmask)
```

```python
# Establish a gate filter
gatefilter = pyart.correct.GateFilter(r2)
#help(gatefilter)
gatefilter.exclude_below('reflectivity', 25)
print("True gates:\n"
      "Original = %d\n" 
      "Excluded = %d\n"
      "NewRadar = %d"%(np.sum(origmask), np.sum(gatefilter.gate_excluded), 
                      np.sum(r2.fields['reflectivity']['data'].mask)))
origmask.shape, gatefilter.gate_excluded.shape
```

```python
#fig, (ax1, ax2) = plt.subplots(1,2, figsize=(13,5))
d2 = pyart.graph.RadarDisplay(r2)
d2.plot('reflectivity', 1, cmap='pyart_LangRainbow12', vmin=-10., vmax=65.,
        gatefilter=gatefilter.gate_excluded)
d2.set_limits((-250., 250), (-250, 250))
```
